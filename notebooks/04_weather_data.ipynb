```python
from datetime import date, timedelta

current_date = date.today()
start_date = date(2012, 12, 3)
dates = []

while start_date <= current_date:
    dates.append(start_date.strftime("%Y-%m-%d"))
    start_date += timedelta(days=1)

print(dates[:3])
```

```python
import requests
from bs4 import BeautifulSoup

weather_data = []

for date in dates:
    url = f'https://www.almanac.com/weather/history/WA/Seattle/{date}'
    response = requests.get(url)
    html_content = response.text
    
    soup = BeautifulSoup(html_content, 'html.parser')
    
    def get_value(row_class, cell_index=0, default='NA'):
        row = soup.find('tr', class_=row_class)
        if row:
            cell = row.find_all('td')[cell_index]
            value_span = cell.find('span', class_='value')
            if value_span:
                return value_span.text.strip()
        return default
    
    # Minimum Temperature
    temp_min = get_value('weatherhistory_results_datavalue temp_mn')
    
    # Maximum Temperature
    temp_max = get_value('weatherhistory_results_datavalue temp_mx')
    
    # Average Temperature
    temp_avg = get_value('weatherhistory_results_datavalue temp')
    
    # Precipitation
    precipitation = get_value('weatherhistory_results_datavalue prcp')
    
    # Average Wind Speed
    wind_speed_avg = get_value('weatherhistory_results_datavalue wdsp')
    
    data = {
        'date': date,
        'temp_min': temp_min,
        'temp_max': temp_max,
        'temp_avg': temp_avg,
        'precipitation': precipitation,
        'wind_speed_avg': wind_speed_avg
    }
    weather_data.append(data)
    print(data)
```

```python
import json
import os

os.chdir("/Users/ayatohisanaga/University of Washington/2024 Spring/IMT 542/project/outputs/weather")

with open("weather_data.json", "w") as outfile:
    json.dump(weather_data, outfile)
    
print("weather data saved successfully.")
```
