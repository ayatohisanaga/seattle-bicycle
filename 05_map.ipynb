```python
map_center = [47.61, -122.3]
seattle_map = folium.Map(location=map_center, zoom_start=10.5)
```

```python
for place, coord in coordinates.items():
    folium.Marker(
        location=coord,
        popup=place,
        tooltip=place
    ).add_to(seattle_map)

# Save the map to an HTML file
seattle_map.save("website/templates/seattle_map.html")
```
